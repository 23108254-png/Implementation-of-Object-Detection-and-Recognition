```python
!pip install ultralytics
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
import requests

# 2. Download a sample image if not already present
image_url = 'https://ultralytics.com/images/bus.jpg'
image_path = 'bus.jpg'

if not os.path.exists(image_path):
    print(f"Downloading {image_path}...")
    response = requests.get(image_url, stream=True)
    response.raise_for_status() # Raise an exception for HTTP errors
    with open(image_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print("Download complete.")
else:
    print(f"{image_path} already exists.")


# 3. Load the pre-trained YOLOv8 model (small version)
model = YOLO('yolov8n.pt') # we can use yolov8s.pt, yolov8m.pt, etc.
# 4. Load your image
image = cv2.imread(image_path)

# Check if image was loaded successfully and has valid dimensions
if image is None or image.shape[0] == 0 or image.shape[1] == 0:
    print(f"Error: Unable to load image from {image_path} or image is empty/invalid.")
else:
    # Convert image from BGR to RGB (for correct color display with matplotlib)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # 5. Perform object detection
    results = model(image_path)
    # 6. Visualize detection results
    # results[0].plot() returns an image with bounding boxes drawn
    annotated_image = results[0].plot()
    # 7. Show the image
    plt.figure(figsize=(12, 8))
    # Convert BGR (OpenCV default) to RGB for matplotlib display
    plt.imshow(annotated_image[..., ::-1])
    plt.axis('off')
    plt.title("Object Detection using YOLOv8")
    plt.show()
```
